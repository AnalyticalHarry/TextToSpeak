### 1. CODE FOR CONVERTING TEXT INTO AUDIO BOOKS  (Pyttsx3)

```python
def speech(text):
        engine = pyttsx3.init() # object creation

        """ RATE"""
        rate = engine.getProperty('rate')   # getting details of current speaking rate
        print()
        print(f"Speed rate of speech: {rate}")  #printing current voice rate
        engine.setProperty('rate', 180)     # setting up new voice rate


        """VOLUME"""
        volume = engine.getProperty('volume')   #getting to know current volume level (min=0 and max=1)
        print()
        print(f"Volume of speech: {volume}")                          #printing current volume level
        engine.setProperty('volume',1.0)    # setting up volume level  between 0 and 1
        engine.say(text)
        engine.runAndWait()
        engine.stop()

        """Saving Voice to a file"""
        # On linux make sure that 'espeak' and 'ffmpeg' are installed
        engine.save_to_file(text, 'test.mp3')
        engine.runAndWait()
```

```python
text = ''' Machine learning is a subfield of artificial intelligence, 
which is broadly defined as the capability of a machine to imitate intelligent human behavior. 
Artificial intelligence systems are used to perform complex tasks in a way that is similar to how humans solve problems.'''
```

```python
speech(text)
```

```python
def audio_book(text, slow=1, lang='en'):
    tts = gTTS(text=text, lang=lang, slow=slow)
    audio_output = tts.save("audio.mp3")
    return audio_output
```

### 3. Converting into FLAC

LAC is a digital audio format that uses lossless compression, meaning audio is compressed without any loss in quality. It stands for Free Lossless Audio Codec and was developed by the Xiph.Org Foundation. It's similar to MP3 but provides better quality as there is no loss in audio during compression.

```python
# Write out audio as 24bit flac, change accordingly
sf.write('audio.flac', audio_data, sample_rate, format='flac', subtype='PCM_24')
```

```python
audio_data, sample_rate = librosa.load('audio.flac', sr=None)
Audio(audio_data, rate=sample_rate)
```
